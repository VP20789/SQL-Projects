# Solving questions
## Question 1: Altona Property Price Hypothesis Test(1 sample T-test)
- Hypothesize that median price in Altona is $800,000.
- Test the sample median/mean against this using a one-sample t-test.

- Alpha=0.05
- H0 = Typical/Median property in suburb of Altona price is $800,000

- H1 = Typical/Median pproperty price has changed/increased than $800,000

```python
alpha=0.05
altona_price=property_df[property_df['Suburb']=='Altona']['Price'].dropna()
altona_price
# Null hypothesis to check mean price = 800000
```

```python
t_test,p_value=stats.ttest_1samp(altona_price,800000)
print('Mean price: ',np.mean(altona_price))
print(t_test)
print(p_value)
```

```python
if p_value<alpha:
    print('Reject H0: Typical/Median property in suburb of Altona price is $800,000')
else:
    print('Fail to reject H0: Typical/Median pproperty price has changed/increased than $800,000')
```

## Question 2: Summer vs. Winter prices for the Year(2016)
- Significance level is 5% i.e. alpha=0.05

- Using T-test for mean

- H0: Significance difference in prices between Summer vs. Winter

- H1: No difference in proces betweeen Summer vs. Winter

```python
alpha=0.05
property_df['Date']=pd.to_datetime(property_df['Date'],format='mixed')
#changing column datatype to Date
```

```python
property_df['Date']
```

```python
date_2016=property_df[property_df['Date'].dt.year == 2016]
date_2016
```

```python
summer_months = [4, 5, 6, 7, 8, 9]
winter_months = [10, 11, 12, 1, 2, 3]

summer_prices = date_2016[date_2016['Date'].dt.month.isin(summer_months)]['Price']
print("Average Summer prices: ", summer_prices.mean())
winter_prices = date_2016[date_2016['Date'].dt.month.isin(winter_months)]['Price']
print("Average Winter prices: ", winter_prices.mean())
```

```python
# Perform t-test to check for significant difference
t_stat, p_value = stats.ttest_ind(summer_prices, winter_prices)
print(t_stat)
print(p_value)
```

```python
if p_value<alpha:
    print("Prices are significantly different between summer & winter")
else:
    print("No significant difference in Prices different between summer & winter")
```
