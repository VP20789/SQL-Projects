## 5. Convert the use_chip column to have only three values: swipe, chip, or online

```python
# map to 2 categories by user defined function
def clean_usechip(val):
    if isinstance(val, str):
        val = val.lower()
        if "online" in val: return "online"
        if "chip" in val: return "chip"
        if "swipe" in val or "magstripe" in val: return "swipe"
    return "swipe" # Default
transactions['use_chip'] = transactions['use_chip'].apply(clean_usechip)
```

## 6. How can you clean up city names, so they don‚Äôt have extra spaces and always start with capital letters?

```python
# using Title() function to ensure capital letter of a word & Strip() function to remove extra spaces
transactions['merchant_city']=transactions['merchant_city'].str.strip().str.title()
```

## 7. Make sure state codes are always two capital letters (like CA, NY)?

```python
#using upper() function to make all letter capital
transactions['merchant_state']=transactions['merchant_state'].str.strip().str.upper()
```

## 8. How do you turn ZIP codes into a proper 5-digit string, keeping leading zeros?

```python
transactions['zip'] = transactions['zip'].astype(str).str.zfill(5)
```

## 3. Customer Spend Profile vs. Credit Health
### ‚Ä¢ For each client_id, compute: total spend, average ticket size, transaction count, and monthly frequency.
### ‚Ä¢ Create a dataframe for credit_score, yearly_income, total_debt, num_credit_cards.
### ‚Ä¢ Visualize their co-operation.
### ‚Ä¢ Plot Scatter of average amount spent vs yearly_income.

```python
# ‚Ä¢ For each client_id, compute: total spend, average ticket size, transaction count, and monthly frequency
client_stats = transactions.groupby('client_id')['amount'].agg(['sum','count'])

# Assuming transaction date is in a month-year format or can be converted to month
transactions['month'] = pd.to_datetime(transactions['date'])
#transactions['month'] = transactions['date'].dt.to_period('M')
monthly_freq = transactions.groupby(['client_id', 'month']).size().groupby('client_id').mean()
```

```python
client_stats['monthly_frequency'] = monthly_freq
client_stats = client_stats.rename(columns={'sum':'total_spend','mean':'avg_ticket'})
client_stats
```

```python
# Merge with credit info
credit_cols = ['client_id','credit_score','yearly_income','total_debt','num_credit_cards']
merged = pd.merge(client_stats.reset_index(), users[credit_cols], on='client_id')
```

```python
# Scatter plot: average spend vs yearly income
plt.figure(figsize=(8,6))
sns.scatterplot(x="avg_ticket", y="yearly_income", data=merged)
plt.title("Average Ticket Size vs Yearly Income")
plt.show()
```

## 4. Age Portfolio:
### ‚Ä¢ Create appropriate age bands. What share of customers falls in each age band? How does the average credit_limit vary by band?
### ‚Ä¢ Visualize age vs chip_usage, income and credit score.
### ‚Ä¢ Extract insights and recommend actions.
### ‚Ä¢ Are young customers online heavy?

```python
age_bins = [0, 25, 35, 50, 65, 100]
age_labels = ["<25","25-35","35-50","50-65","65+"]
users['age_band'] = pd.cut(users['current_age'], bins=age_bins, labels=age_labels)
```

```python
age_share = users['age_band'].value_counts(normalize=True)
plt.figure(figsize=(8,6))
age_share.plot(kind="bar")
plt.title("Customer Share by Age Band")
plt.show()
```

```python
# Avg. credit limit by band
age_card_merge = pd.merge(users[['id','age_band']], cards[['id','credit_limit']])
avg_creditlimit_by_band = age_card_merge.groupby('age_band')['credit_limit'].mean()
plt.figure(figsize=(8,6))
avg_creditlimit_by_band.plot(kind="bar")
plt.title("Average Credit Limit by Age Band")
plt.show()
```

```python
# ‚Ä¢ Visualize age vs chip_usage, income and credit score.
plt.figure(figsize=(8,6))
sns.boxplot(x="age_band", y="credit_score", data=users)
plt.title("Credit Score by Age Band")
plt.show()
```

```python
# Income by age band
plt.figure(figsize=(8,6))
sns.boxplot(x="age_band", y="yearly_income", data=users)
plt.title("Income by Age Band")
plt.show()
```

```python
# ‚Ä¢ Are young customers online heavy
young_clients = users.loc[users['age_band'] == "<25", "id"]
young_online = transactions[(transactions['client_id'].isin(young_clients))]
print("Share of online transactions by clients <25:", len(young_online) / len(transactions['client_id']))
```
